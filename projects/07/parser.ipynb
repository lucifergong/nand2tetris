```python
def codewriter():
    def __init__(self,filename):
        filename=filename.split('.')
        self.current_translated_file_name=filename[0]
        self.label_num = 0
        self.current_function_name=0
        self.if_in_function=0
        self.return_label_num=0
    def push(self,q):
        if q[1]=='constant':
            write_codes(['@'+str(int(q[2])),'D=A'])
            self.write_push_from_d_register
        elif q[1] in  ["local", "argument", "this", "that"]:
            self.write_push_from_virtual_segment(segment,q[1],q[2])
        elif q[1] in ["temp", "pointer"]:
            self.write_push_from_static_segment(segment,q[1],q[2])
        elif q[1] == "static":
                self.write_codes([
                    "@%s.%d" % (self.current_translated_file_name,q[2]),
                ])
                self.write_code('D=M')
                self.write_push_from_d_register()
            
    def write_codes(self,w):
        for i in w:
            f.write((i+'\n'))
    def write_push_from_d_register(self):
        self.write_codes([
            '@SP',
            'A=M',
            'M=D',
            '@SP',
            'M=M+1'
        ])
    def write_push_from_virtual_segment(self,c,d):
        if c=="local":
            virtual_address="LCL"
        elif c =="argument":
            virtual_address="ARG"
        elif c=="this":
            virtual_address="THIS"
        elif c=="that":
            virtual_address="THAT"
        self.write_codes([
            '@%s' % virtual_address,
            'A=M'
        ])
        for i in range(d):
            self.write_codes(['A=A+1'])
        self.write_codes(['D=A'])  #存疑，难道不该是把这个地址中存的M取出来吗？
        self.write_push_from_d_register
    def write_push_from_static_segment(self,segment,f):
        if segment == "temp":
            real_address = TEMP_BASE_ADDRESS
        elif segment == "pointer":
            real_address = POINTER_BASE_ADDRESS
        self.write_codes([
            '@%s' % real_address,
        ])
        for i in range(d):
            self.write_codes(['A=A+1'])
        self.write_codes(['D=A'])
        self.write_push_from_d_register
        
        
    def point_up(self):
        write_code(['@SP',
                    'M=M-1',
                    'A=M'
        ])
    def pop(self,q):
        if q[1] in  ["local", "argument", "this", "that"]:
            self.write_pop_from_virtual_segment(segment,q[1],q[2])
        elif q[1] in ["temp", "pointer"]:
            self.write_pop_from_static_segment(segment,q[1],q[2])
        elif q[1] == "static":
                self.point_up
                write_codes(['D=M'])
                self.write_codes([
                    "@%s.%d" % (self.current_translated_file_name,q[2]),
                ])
                self.write_code('D=M')
    
    def write_pop_from_virtual_segment(self,c,d):
        self.point_up()
        write_code(['D=M'])
        if c=="local":
            virtual_address="LCL"
        elif c =="argument":
            virtual_address="ARG"
        elif c=="this":
            virtual_address="THIS"
        elif c=="that":
            virtual_address="THAT"
        self.write_codes([
            '@%s' % virtual_address,
            'A=M'
        ])
        for i in range(d):
            self.write_codes(['A=A+1'])
        self.write_codes(['M=D'])
    def write_pop_from_static_segment(self,c,d):
        self.point_up()
        write_codes(['D=M'])
        if c == "temp":
            real_address = TEMP_BASE_ADDRESS
        elif c == "pointer":
            real_address = POINTER_BASE_ADDRESS
        self.write_codes([
            '@%s' % virtual_address,
        ])
        for i in range(d):
            self.write_codes(['A=A+1'])
        self.write_codes(['M=D'])

        
        
    def write_arithmetic(self, command):
        if command in ["add", "sub", "and", "or"]:
            self. if q=='add':(command)
        elif command in ["neg", "not"]:
            self.write_unary_operation(command)
        elif command in ["eq", "gt", "lt"]:
            self.write_comp_operation(command)
    def write_binary_operation(self,command):
        self.point_up()
        self.write_codes(['D=M'])
        self.point_up()
        if command == 'add':
            self.write_codes('D=D+M')
        elif command == 'sub':
            self.write_codes('D=M-D')
        elif command == 'and':
            self.write_codes('D=D&M')
        elif command == 'or':
            self.write_codes('D=D|M')
        self.write_push_from_d_register()
    def write_unary_operation(self, command):
        self.write_codes([
            '@SP',
            'A=M-1',
        ])
        if command == 'neg':
            self.write_code('M=-M')
        elif command == 'not':
            self.write_code('M=!M')
            
    def write_comp_operation(self, command):
        self.point_up()
        self.write_codes('D=M')
        self.point_up()
        l1 = self.get_new_if_label()
        l2 = self.get_new_if_label()
        if command == "eq":
            comp_type = "JEQ"
        elif command == "gt":
            comp_type = "JGT"
        elif command == "lt":
            comp_type = "JLT"
        self.write_codes([
            'D=M-D',
            "@%s" % l1,
            'D;%s' % comp_type,
            'D=0',
            "@%s" % l2,
            '0;JMP',
            "(%s)" % l1,
            'D=-1',
            "(%s)" % l2,
        ])
        self.write_push_from_d_register()

        
    def get_new_if_label(self):
        self.if_label_num += 1
        return 'if_LABEL' + str(self.if_label_num)
    def write_bootstrap(self):
        self.write_codes(['@256',
                     'D=A',
                     '@SP',
                     'M=D'])
        self.write_call("Sys.init", 0)
# unfinished
#    def write_Sys_init(self):
#        l3 = self.get_new_label()
 #       self.write_codes(['call main',
  #                   "(%s)" % l3,
#                     '0;JMP'])
    
    
    def get_real_label_name(self,label):
        if (self.if_in_function):
            real_label=( "("+self.current_function_name + label +")" ) 
        else:
            real_label=label
        return label
        
    def write_label(self,w):
        w=w.split(' ')
        label=self.get_real_label_name(w[1])
        self.write_codes([label])
    
    def write_goto(self,w):
        w=w.split(' ')
        label=get_real_label_name(w[1])
        self.write_codes(['@%s' %label,
                        '0,JMP' ])
    
    def write_if(self,q):
        q=q.split(' ')
        self.point_up()
        self.write_codes(['D=M',
                        '@%s' %self.get_real_label_name(q[1]),
                        'D;JNE'])
        
    def get_new_return_label(self):
        self.return_label_num += 1
        return '_RETURN_LABEL_' + str(self.return_label_num)        
    
        
    def write_call(self,function,n):
        self.current_function_name=function
        return_label = self.get_new_return_label()
        
        
        
        self.write_codes([
            '// return-label',
            '@%s' % return_label,
            'D=A'
        ])        
        self.write_push_from_d_register()  # push return-address
        
        
        
        self.write_codes([
            '@LCL',
            'D=M',
        ])
        self.write_push_from_d_register()
        
        
        
        
        self.write_codes([
            '@ARG',
            'D=M',
        ])
        self.write_push_from_d_register()  
        self.write_codes([
            '@THIS',
            'D=M',
        ])
        self.write_push_from_d_register()  
        self.write_codes([
            '@THAT',
            'D=M',
        ])
        self.write_push_from_d_register()
        write_codes(['@SP'])
        for i in range(n+5):
            write_codes(['A=A-1'])  #可能不对
        write_codes(['D=A',
                     '@ARG',
                     'M=D',
                      '@SP',
                      'D=M',
                      '@LCL',
                      'M=D'])
        fun='goto %' %self.current_function_name
        self.write_goto('goto %')
        self,write_codes([return_label])
        
    def write_function(self,f,k):
        self.current_function_name=f
        self.if_in_function=1
        self.write_codes([
            '(%s)' % function_name,
            'D=0'
        ])
        for i in range(int(k)):
            self.write_push_from_d_register()
    def write_return(self):
        self.write_codes(['@LCL',
                          'D=M',
                          '@R13',
                          'M=D',
                          '@5',
                          'D=A',
                          '@R13',
                          'A=M-D',
                          'D=M',
                          '@R14',
                          'M=D',  # R14 = return-address     
                          ])
        self.point_up()
        self.write_codes([
            'D=M',
            '@ARG',
            'A=M',  # M = *ARG
            'M=D',  # *ARG = pop()

            '@ARG',
            'D=M+1',
            '@SP',
            'M=D',  # SP = ARG + 1

            '@R13',
            'AM=M-1',  # A = FRAME-1, R13 = FRAME-1
            'D=M',
            '@THAT',
            'M=D',  # THAT = *(FRAME-1)

            '@R13',
            'AM=M-1',
            'D=M',
            '@THIS',
            'M=D',  # THIS = *(FRAME-2)

            '@R13',
            'AM=M-1',
            'D=M',
            '@ARG',
            'M=D',  # ARG = *(FRAME-3)

            '@R13',
            'AM=M-1',
            'D=M',
            '@LCL',
            'M=D',  # LCL = *(FRAME-4)

            '@R14',
            'A=M',
            '0;JMP'  # goto return-address
        ])
```
